```python
import sys
import os
sys.path.append(os.path.abspath(".."))
%load_ext autoreload
%autoreload 2
from src.model import train_model
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df_train = pd.read_parquet('../data/train_data/df_train_final.parquet')
df_test = pd.read_parquet('../data/train_data/df_test_final.parquet')
```

```python
import re

def clean_feature_names(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    cleaned_cols = []

    for col in df.columns:
        col = str(col)

        # replace any non-alphanumeric character with underscore
        col = re.sub(r"[^A-Za-z0-9_]+", "_", col)

        # collapse repeated underscores
        col = re.sub(r"_+", "_", col)

        # remove leading/trailing underscores
        col = col.strip("_")

        # LightGBM also dislikes empty names
        if col == "":
            col = "EMPTY_COL"

        cleaned_cols.append(col)

    # ensure uniqueness after cleaning
    counts = {}
    unique_cols = []
    for col in cleaned_cols:
        if col in counts:
            counts[col] += 1
            unique_cols.append(f"{col}_{counts[col]}")
        else:
            counts[col] = 0
            unique_cols.append(col)

    df.columns = unique_cols
    return df



result = train_model(clean_feature_names(df_train), clean_feature_names(df_test))
```
